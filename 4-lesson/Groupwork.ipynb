```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('earthquake_data_tsunami.csv')
df
```

```python
# Data types
df.info()
```

```python
# Visualising missing data
df.isnull().sum()
```

```python
# Visualizing the integers here
missing = df.isnull().astype(int)
plt.figure(figsize=(10,6))
plt.imshow(missing, aspect='auto', cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Missing Value (1 = Missing, 0 = Present)')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Missing Values Heatmap')
plt.xticks(rotation=90)
```

```python
# Visualizing the floats here
missing = df.isnull().astype(float)
plt.figure(figsize=(10,6))
plt.imshow(missing, aspect='auto', cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Missing Value (1 = Missing, 0 = Present)')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Missing Values Heatmap')
plt.xticks(rotation=90)
```

```python
df_corr = df.corr()
```

```python
# Visualizing using heatmap via Seaborn

plt.figure(figsize=(12, 8))
sns.heatmap(
    df_corr,
    annot=True,              # Show correlation values
    fmt=".2f",               # Format to 2 decimal places
    cmap="Blues",            # Color palette for contrast
    vmin=-1, vmax=1,         # Fix scale for consistency
    linewidths=0.5,          # Add lines between cells
    linecolor='gray',        # Line color
    square=True,             # Make cells square
    cbar_kws={"shrink": 0.8} # Shrink color bar
)
plt.title("Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
```
