```python
df['magnitude'].plot.kde()
```

```python
plt.figure(figsize=(8,6))
img = plt.imread('map.png')
plt.imshow(img, extent=[126, 130, 34, 38], origin='upper')
plt.scatter(df['longitude'], df['latitude'], s=40, alpha=0.2, c='r')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Data Points')
```

```python
plt.figure(figsize=(20, 15))
img = plt.imread('map.png')

# Dynamic extent based on data
lon_min, lon_max = df['longitude'].min(), df['longitude'].max()
lat_min, lat_max = df['latitude'].min(), df['latitude'].max()

# Add a little margin around points
lon_pad = (lon_max - lon_min) * 0.1
lat_pad = (lat_max - lat_min) * 0.1

plt.imshow(img, extent=[lon_min - lon_pad, lon_max + lon_pad,
                        lat_min - lat_pad, lat_max + lat_pad])

plt.scatter(df['longitude'], df['latitude'], s=150, alpha=0.1, c='r')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Data Points')
plt.show()
```

```python
df_corr=df.corr()
```

```python
sorted_corr = df_corr['tsunami'].abs().sort_values(ascending=False)
sorted_corr
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(
    df_corr,
    annot=True,              # Show correlation values
    fmt=".2f",               # Format to 2 decimal places
    cmap="Blues",         # Color palette for contrast
    vmin=-1, vmax=1,         # Fix scale for consistency
    linewidths=0.5,          # Add lines between cells
    linecolor='gray',        # Line color
    square=True,             # Make cells square
    cbar_kws={"shrink": 0.8} # Shrink color bar
)
plt.title("Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
```

```python
tsunami_counts = df[df['tsunami']==1]
tsunami_counts
```

```python
tsunami_counts_sorted_by_year = tsunami_counts.groupby('Year').size()
tsunami_counts_sorted_by_year
```

```python
plt.figure(figsize=(12,6))
tsunami_counts_sorted_by_year.plot(kind='bar', color='skyblue')
plt.title('Tsunami Events per Year')
plt.xlabel('Year')
plt.ylabel('Tsunami Counts')
plt.xticks(rotation=45)
plt.tight_layout()
```
