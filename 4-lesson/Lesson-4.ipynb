```python
plt.figure(figsize=(20, 15))
img = plt.imread('map.png')

# Dynamic extent based on data
lon_min, lon_max = df['longitude'].min(), df['longitude'].max()
lat_min, lat_max = df['latitude'].min(), df['latitude'].max()

# Add a little margin around points
lon_pad = (lon_max - lon_min) * 0.1
lat_pad = (lat_max - lat_min) * 0.1

plt.imshow(img, extent=[lon_min - lon_pad, lon_max + lon_pad,
                        lat_min - lat_pad, lat_max + lat_pad])

plt.scatter(df['longitude'], df['latitude'], s=150, alpha=0.1, c='r')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Data Points')
plt.show()
```

```python
df_corr=df.corr()
```

```python
sorted_corr = df_corr['tsunami'].abs().sort_values(ascending=False)
sorted_corr
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(
    df_corr,
    annot=True,              # Show correlation values
    fmt=".2f",               # Format to 2 decimal places
    cmap="Blues",         # Color palette for contrast
    vmin=-1, vmax=1,         # Fix scale for consistency
    linewidths=0.5,          # Add lines between cells
    linecolor='gray',        # Line color
    square=True,             # Make cells square
    cbar_kws={"shrink": 0.8} # Shrink color bar
)
plt.title("Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
```

```python
df2 = pd.read_csv('air_quality_global.csv')
df2
```

```python
df2.drop(['latitude', 'longitude'], axis=1)
```

```python
usa_reg = df2[df2['country'] == 'USA']
usa_reg_good_quality = usa_reg[usa_reg['data_quality'] != 'Poor']
usa_reg_good_quality
```

```python
usa_reg_good_quality.info()
```

```python
usa_reg['year'] = pd.to_numeric(df2['year'], errors='coerce')
```

```python
city_year_avg = usa_reg_good_quality.groupby(['city', 'year'])['pm25_ugm3'].mean().reset_index()
city_year_avg
```

```python
plt.figure(figsize=(12,6))
cities = city_year_avg['city'].unique()

for city in cities:
    city_data = city_year_avg[city_year_avg['city'] == city]
    plt.plot(city_data['year'], city_data['pm25_ugm3'], marker='o', label=city)

plt.xlabel('Year')
plt.ylabel('PM2.5 (Âµg/mÂ³)')
plt.title('PM2.5 Levels per Year per City')
plt.legend()
plt.grid(True)
```

```python
for city in cities:
    city_data = usa_reg_good_quality[usa_reg_good_quality['city'] == city].groupby('year')['pm25_ugm3'].mean().reset_index()
    
    plt.figure(figsize=(8,5))
    plt.plot(city_data['year'], city_data['pm25_ugm3'], marker='o', linestyle='-', color='b')
    plt.title(f'PM2.5 Levels Over Years in {city}')
    plt.xlabel('Year')
    plt.ylabel('PM2.5 (Âµg/mÂ³)')
    plt.grid(True)
```
